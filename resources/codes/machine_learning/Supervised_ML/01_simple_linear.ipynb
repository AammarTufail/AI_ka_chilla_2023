# Simple Linear Regression model in Python

- Simple linear regression is a statistical method that allows us to summarize and study relationships between `two continuous (quantitative) variables`. 
- One dependent and the other independent variable.
- The goal of simple linear regression is to build a `mathematical equation` that predicts the `dependent variable as a function of the independent variable`. 
- The equation is of the form: `y = mx + b`. 
  - The slope of the line is m, and b is the y-intercept. The independent variable is often called the explanatory variable, and the dependent variable is often called the response variable. 

```python
# general libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#machine learning
from sklearn.linear_model import LinearRegression
```

```python
# step one - load the data
df = sns.load_dataset('tips')
X = df[['total_bill']]
y = df['tip']
```

```python
# step two - create the model
model = LinearRegression()
```

```python
# step three - fit the model (train the model)
model.fit(X, y)
```

```python
#step four - Make prediction
model.predict([[3]])
```

```python
# Let's plot the model
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Let's plot the model
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# slope, intercept, r2_value
print("Intercept: ", model.intercept_)
print("Slope: ", model.coef_)
print("R2: ", model.score(X, y))
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm


# Fit a linear regression model
model = sm.OLS(y, sm.add_constant(X)).fit()

# Get the p-value of the slope coefficient
p_value = model.params[1]
print("p-value:", p_value)
```

# SLR on scaled data


```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# load the data
df = sns.load_dataset('tips')
X = df[['total_bill']]
y = df['tip']

#scale X
# scalar = MinMaxScaler()
scalar = StandardScaler()
X_scaled = scalar.fit(X)

# transform data
X_scaled = scalar.transform(X)

#create a model
model = LinearRegression()

#train a model
model.fit(X_scaled, y)
```

```python
# slope, intercept, r2_value
print("Intercept: ", model.intercept_)
print("Slope: ", model.coef_)
print("R2: ", model.score(X_scaled, y))
```

```python
# slope, intercept, r2_value
print("Intercept: ", model.intercept_)
print("Slope: ", model.coef_)
print("R2: ", model.score(X_scaled, y))
```
