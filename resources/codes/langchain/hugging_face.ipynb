# How to use Hugging Face API for GPT like Tasks

```python
# Import LangChain and Hugging Face pipeline
from langchain import OpenAI
from transformers import pipeline

# Initialize Hugging Face summarization pipeline
summarizer = pipeline("summarization")

# Initialize LangChain with Hugging Face model
ai = OpenAI(model_name="text-davinci-003")

# Get text to summarize  
text = "The Hubble Space Telescope has captured a stunning image of the spiral galaxy NGC 2008, located around 425 million light-years away in the constellation Pictor. The galaxy is oriented nearly face-on as seen from Earth, allowing astronomers to study the intricate spiral structure of its disc and the bright knots of newborn stars."

# Summarize text using Hugging Face pipeline
hf_summary = summarizer(text)[0]['summary_text']

# Summarize using LangChain + Hugging Face 
lc_summary = ai.generate(
  prompt = f"Summarize this text:\n\n{text}\n\nSummary:", 
  model="text-davinci-003"
)

print(hf_summary)
print(lc_summary)
```
